```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
hotel_data = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv")
```

# Histogram

```python
plt.figure(figsize = (10,8))

sns.set_style("whitegrid")
sns.histplot(data = hotel_data, binrange = (0,300), x="price", binwidth = 12,legend = False, alpha=0.8)
sns.despine(left = True)

plt.xlabel("Price")
plt.title("Price distribution")

plt.show()
```

# Density Plot

```python
plt.figure(figsize = (8,5))
hotel_data = hotel_data[hotel_data['price'] < 300]

sns.kdeplot(data = hotel_data, x="price",fill = True, alpha=0.9)

sns.despine()

plt.xlabel("Price")

plt.show()
```

# Story 6

```python
url = 'https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/5_OneCatSevNumOrdered.csv'
df = pd.read_csv(url)
```

```python
data = df[(df['name'].isin(["Ashley", "Amanda", "Jessica", "Patricia", "Linda", "Deborah", "Dorothy", "Betty", "Helen"])) & (df['sex'] == 'F')]
data = data[['year','name','n']]
data
```

### Line Plot

```python
plt.figure(figsize=(12,6))
for name in data.name.unique():
    d = data[data['name'] == name]
    plt.plot(d['year'], d['n'], label=name)
plt.legend(loc='upper right')
plt.title("Line Plot")
plt.xlabel("Year")
plt.ylabel("Number of babies")
plt.show()
```

### Area Plot

```python
plt.figure(figsize=(10,8))
for name in data.name.unique():
    d_name = data[data['name'] == name]
    plt.fill_between(d_name['year'], d_name['n'], label=name, alpha=0.5)

plt.title("Area Plot")
plt.xlabel("Year")
plt.ylabel("N0. of babies")
plt.legend(loc='upper right', bbox_to_anchor=(1, 1), ncol=1)
plt.show()
```

```python
df
```

```python
data = df[['disp', 'drat', 'hp', 'mpg', 'qsec', 'wt']]
data = data.melt()
data
```

### Multiple Density Plot

```python
data = df[['mpg', 'disp', 'drat', 'hp', 'qsec', 'wt']]

sns.pairplot(data, kind='scatter', diag_kind='kde', markers='.', plot_kws={'alpha': 0.5, 's': 30, 'color': '#69b3a2'}, diag_kws={'color': '#69b3a2'})
plt.title('Multiple Density Plot')
plt.show()
```

### Heat Map

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data, cmap='coolwarm', annot=False, fmt='.2f')
```

```python
from wordcloud import WordCloud
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/7_OneCatOneNum.csv")
data
```

```python
mywords = data.groupby('Country')['Value'].count().sort_values(ascending=False)[:50]
```

```python
wordcloud = WordCloud(width = 800, height = 800, background_color ='white',
                      color_func = lambda *args, **kwargs: "#69b3a2",
                      min_font_size = 10).generate_from_frequencies(mywords)

plt.figure(figsize = (10, 10), facecolor = 'white')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
